```python
import numpy as np
import pandas as pd
```

```python
%cd ..
from src.fetch import fetch_inspection_data

df = fetch_inspection_data()
```

```python
df = df.loc[pd.to_numeric(df['inspection_year']) != 1900]
```

```python
print("Positivity Rate:",len(df.loc[df['is_closed'] == 1]) / len(df))
```

```python
df.columns
```

```python
# Closures by date analysis
date_closures = df[['inspection_year', 'inspection_month', 'is_closed']]
date_closures = date_closures.groupby(['inspection_year', 'inspection_month']).sum()
date_closures = date_closures.sort_values(['inspection_year', 'inspection_month'])
```

```python
%matplotlib inline
date_closures.hist()
```

```python
date_closures.plot()
```
